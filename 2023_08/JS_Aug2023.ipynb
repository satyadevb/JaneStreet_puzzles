```python
import math
import numpy as np
import matplotlib.pyplot as plt

N = 10000
```

```python
def prob3(D,N):
    p = 0
    for i in range(N):
        r = np.random.uniform(0,1)
        theta = 2*np.arcsin(np.sqrt(r))
        phi = np.random.uniform(0,2*math.pi)
        x = np.random.uniform(0,1)
        y = np.random.uniform(0,1)
        z = np.random.uniform(0,1)
        lx = D*math.sin(theta)*math.cos(phi)
        ly = D*math.sin(theta)*math.sin(phi)
        lz = D*math.cos(theta)
        crosses = abs(math.floor(x + lx) - math.floor(x)) + abs(math.floor(y + ly) - math.floor(y)) + abs(math.floor(z + lz) - math.floor(z))
        if( crosses == 1 ):
            p = p + 1
    return p/N

D_list = np.linspace(0,1,100)
prob = np.array([])

for i,D in enumerate(D_list):
    if i % 100 == 0:
        print(i)
    prob = np.append(prob,prob3(D,N))
    
plt.plot(D_list,prob)

plt.show()
```

```python
for i in range(10):
    l = []
    for i in np.linspace(0,1,100):
        l.append(prob3(i,10000))
    plt.plot(np.linspace(0,1,100),l)
plt.show()
```
